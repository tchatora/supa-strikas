```python
import pandas as pd
import altair as alt
```

```python
data = pd.read_csv('~/my-book/data/transactions_2025_03_16.csv')
```

```python
data_claims = data[data.loc[:,'Type'] == 'Drop'].reset_index(drop = True)
data_claims.head()
```

```python
most_claims = data_claims['Team.1'].value_counts().reset_index()
most_claims = most_claims.rename(columns = {'index':'Name'}).astype(str)
#most_claims = most_claims.assign(Claims = most_claims['Team.1'].astype(int))
#most_claims.info()
most_claims.head()
```

```python
plot_most = alt.Chart(most_claims, height = 300, width = 500
                     ).mark_bar().encode(
    x = alt.X('Name:N', sort = '-y'), 
    y = alt.Y('Claims:Q')).properties()

plot_most
```
